```python
from keras.preprocessing import image
```

```python
img=image.load_img("cat.jpg",target_size=(500,500))
```

```python
img.size
```

```python
X= image.img_to_array(img)
```

```python
X=X/255.0
```

```python
import matplotlib.pyplot as plt
```

```python
plt.axis("off")
plt.imshow(X)
plt.show()
```

```python
X=X.reshape((1,500,500,3)) 
## keras data generator accepts a batch of image
```

```python
data_generator=image.ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    shear_range=0.2,
    zoom_range=0.2,
    )
```

```python
i=0
for batch in data_generator.flow(X,batch_size=1): # flow is a iterator
    plt.figure()
    plt.axis("off")
    print(i)
    plt.imshow(image.img_to_array(batch[0]))
    i+=1
    if(i==10):
        break
    plt.show()
```
